```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df = pd.read_csv('/content/IOT_Devices_data.xlsx - Sheet1.csv')
df.head()
```

```python
df['Maintenance Required'].value_counts()
```

```python
df_maintenance_required = df[df['Maintenance Required'] == "Yes"]
```

```python
df_new = df[df['Maintenance Required'] == "No"][:100]
```

```python
df_maintenance_required.shape
```

```python
df.head()
```

```python
df_new = pd.concat([df_maintenance_required, df_new])
```

```python
df_new.head()
```

```python
df_new['Maintenance Required'].value_counts()
```

```python
df_new['Maintenance Required'] = le.fit_transform(df_new['Maintenance Required'])
```

```python
df_new_1 = pd.get_dummies(df_new.drop(['Timestamp','Maintenance Required'],axis = 1))
```

```python
df_new_1.head()
```

```python
df_new_1.info()
```

```python

import numpy as np
df_new_1 = df_new_1.astype(np.int64)
df_new_1.head()
```

```python
X = df_new_1.drop(['id'],axis=1)
Y = df_new['Maintenance Required']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
models_list = [LogisticRegression(), SVC(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
for model in models_list:
  model.fit(X_train, y_train)
  y_pred_train = model.predict(X_train)
  score = accuracy_score(y_train, y_pred_train)

  print(f"{model} got {score} accuracy")
```

LogisticRegression() got 0.9640287769784173 accuracy
SVC() got 0.60431654676259 accuracy
RandomForestClassifier() got 1.0 accuracy
KNeighborsClassifier() got 0.7410071942446043 accuracy


so with high train accuracy RandomForestClassifier won

```python
model_2 = RandomForestClassifier()
model_2.fit(X_train, y_train)
Y_pred_test = model_2.predict(X_test)
score = accuracy_score(y_test, Y_pred_test)
print(score)
```
